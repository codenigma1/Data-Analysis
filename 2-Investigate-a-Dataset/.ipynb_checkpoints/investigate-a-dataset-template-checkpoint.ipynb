```python
# I define the fucntion that retrun maximum and minimum value column in term of movie_profit 
def max_min(column):
    
    # obtain max value of column 
    df_max = df.iloc[df[column].idxmax()]
    
    # obtain min value of column
    df_min = df.iloc[df[column].idxmin()]
    
    # concatenate both information
    df_info = pd.concat([df_max, df_min], axis=1)
    
    # return the result
    return df_info
```

### Research Question 3 (Which is the popular movie genres in term of movie profit?)

#### Logic:  
> I imagine like above result **genre_1, genre_2, genre_3, genre_4, genre_5** each hold muliple genres, then calculating movie profit of sum for the indivisual columns. Finally, I combine them altogether and calcuate the sum of their respective genre which is provide best movie genre in term movie profit.

Apply groupby function as sum summary on feature columns, then converted into the dataframe

```python
# Define the function 
def column_group(col_name):
    
    # store the result in variable as form of dataframe 
    col_return = pd.DataFrame(df.groupby(col_name)['movie_profit'].sum())
    
    # return result with ascending order
    return col_return.sort_values(by='movie_profit', ascending=False)
```

I define the function and reduce redundacy of the code. You can see above comments. 

Also, define the **compute function** which is combine the all dataframe, then return _best value_ in term of profit

```python
# defination of function
def compute(columns_pd):
    
    # combine the datafram
    df_concat = pd.concat(columns_pd, ignore_index=False)
    
    # reset the index
    df_concat.reset_index(inplace=True)
    
    # rename the appropriate column
    df_concat.rename(columns={'index': 'name'}, inplace=True)
    
    # Use group function to get summary of each genre
    df_best = df_concat.groupby('name').sum()
    
    # Sort the value by descending order
    df_best_sort = df_best.sort_values(by='movie_profit', ascending=False)
    
    
    # return the final output
    return df_best_sort
```

```python
# Invoke the function of column_group and store result into the variables
df_gen_1 = column_group('genre_1')
df_gen_2 = column_group('genre_2')
df_gen_3 = column_group('genre_3')
df_gen_4 = column_group('genre_4')
df_gen_5 = column_group('genre_5')
```

```python
# it is look like
df_gen_5.head()
```

```python
# Invoke the computer fucntion and store the result
best_genre = compute([df_gen_1,df_gen_2,df_gen_3,df_gen_4,df_gen_5])
```

```python
# series_genre = best_genre.iloc[:,0]
# Showing the result top 5
best_genre.head()
```

**Adventure** is the best genre which has made highest profit, then Action, Comedy, and Drama as subsequent genres

Let's used the Graphical analysis with bar plot

```python
# Creating the horizintal bar graph because seems to be approporiate for view
best_genre.sort_values(by='movie_profit', ascending=True, inplace=True)
best_genre.plot.barh(figsize=(18, 10), rot=15, color='purple', fontsize = '12')

# title and label
plt.title('popular movie genres in term of movie profit', fontsize = '16')
plt.ylabel('Genres', fontsize='16', color = 'black')
plt.xlabel('Amount in one hundred millions Dollar', fontsize='16', color = 'black')
plt.show(block=True)
```

### Research Question 6 (Which is the highest and lowest budget of movie genre?)

```python
# Calculate the alternative ways without function on the basis of the budget
df_genbudget_1 = pd.DataFrame(df.groupby('genre_1')['budget_(US_dollar)'].sum())
df_genbudget_2 = pd.DataFrame(df.groupby('genre_2')['budget_(US_dollar)'].sum())
df_genbudget_3 = pd.DataFrame(df.groupby('genre_3')['budget_(US_dollar)'].sum())
df_genbudget_4 = pd.DataFrame(df.groupby('genre_4')['budget_(US_dollar)'].sum())
df_genbudget_5 = pd.DataFrame(df.groupby('genre_5')['budget_(US_dollar)'].sum())
```

```python
# defination of function
def compute_budget(columns_pd):
    
    # combine the datafram
    df_concat = pd.concat(columns_pd, ignore_index=False)
    
    # reset the index
    df_concat.reset_index(inplace=True)
    
    # rename the appropriate column
    df_concat.rename(columns={'index': 'genre'}, inplace=True)
    
    # Use group function to get summary of each genre
    df_best = df_concat.groupby('genre').sum()
    
    # Sort the value by descending order
    df_best_sort = df_best.sort_values(by='budget_(US_dollar)', ascending=False)
    
    # reset the index
    df_best_sort.reset_index(inplace=True)
    
    # store the result highest and lowest
    highest = df_best_sort.iloc[df_best_sort['budget_(US_dollar)'].idxmax()]
    lowest = df_best_sort.iloc[df_best_sort['budget_(US_dollar)'].idxmin()]
    
    # combine
    final_concat = pd.concat([highest, lowest], axis=1)
    
    # return the final output
    return final_concat
```

```python
gen_budget = compute_budget([df_genbudget_1, df_genbudget_2, df_genbudget_3, df_genbudget_4, df_genbudget_5])
```

```python
gen_budget
```

The movie maker would like to spend dollar on the _**Action**_ genre while  **_TV Movies_** genre has lowest.

### Reaseach Question 7 (Compare the movie genre produce in year 2000 and 2015?)

```python
# Define the fucntion
def combine_year_2000(col_name):
    
    # count genre in year 2000 and store result
    gen_2000 = pd.DataFrame(df[df.release_year == 2000].groupby(col_name)['release_year'].count())
    
    # return result
    return gen_2000

# Similary
def combine_year_2015(col_name):
    
    # count genre in year 2015 and store result
    gen_2015 = pd.DataFrame(df[df.release_year == 2015].groupby(col_name)['release_year'].count())
    
    # return result
    return gen_2015
```

```python
# Year 2000 dataframe
gen1_2000 = combine_year_2000('genre_1')
gen2_2000 = combine_year_2000('genre_2')
gen3_2000 = combine_year_2000('genre_3')
gen4_2000 = combine_year_2000('genre_4')
gen5_2000 = combine_year_2000('genre_5')

# Year 2015 dataframe
gen1_2015 = combine_year_2015('genre_1')
gen2_2015 = combine_year_2015('genre_2')
gen3_2015 = combine_year_2015('genre_3')
gen4_2015 = combine_year_2015('genre_4')
gen5_2015 = combine_year_2015('genre_5')
```

```python
# check and confirm 
gen1_2000
```

```python
# defination of function
def compute_year(columns_pd):
    
    # combine the datafram
    df_concat = pd.concat(columns_pd, ignore_index=False)
    
    # reset the index
    df_concat.reset_index(inplace=True)
    
    # rename the appropriate column
    df_concat.rename(columns={'index': 'name'}, inplace=True)
    
    # Use group function to get summary of each genre
    df_best = df_concat.groupby('name').sum()
    
    # Sort the value by descending order
    df_best_sort = df_best.sort_values(by='release_year', ascending=False)
    
    
    # return the final output
    return df_best_sort
```

```python
# Invoke compute_year function and rename the columns
gen_2000_data = compute_year([gen1_2000,gen2_2000,gen3_2000,gen4_2000,gen5_2000])
gen_2000_data.rename(columns={'release_year':'release_year_(2000)'}, inplace=True)

# Invoke compute_year function and rename the columns
gen_2015_data = compute_year([gen1_2015,gen2_2015,gen3_2015,gen4_2015,gen5_2015])
gen_2015_data.rename(columns={'release_year':'release_year_(2015)'}, inplace=True)
```

```python
# Replace NaN value to 0
comp_2000_2015 = pd.concat([gen_2000_data, gen_2015_data], axis=1).replace(np.nan, 0)
```

```python
# Because of Nan it's converted into the flot. I should fix this datatype into int64
comp_2000_2015= comp_2000_2015.astype({'release_year_(2015)': 'int64'})
```

```python
# Final result of 2000 and 2015 genre production
comp_2000_2015
```

Thus, release year in 2000 almost all genre comparatively less than 2015 except **Foreign, and Western genre.**
